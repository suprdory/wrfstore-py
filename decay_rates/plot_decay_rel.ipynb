```python
from matplotlib import pyplot as plt
import pandas as pd
plt.rcParams['figure.facecolor'] = 'white'
```

```python
# df0w=pd.read_pickle("../rundf/df_dec_run_land_o.pkl")
df1w=pd.read_pickle("../rundf/df_dec_run_land.pkl")
df1d=pd.read_pickle("../rundf/df_dec_run_land_dry.pkl")
```

```python
def plot_dec(ax,x,y,d,xl,yl):
    ax.plot(x,y)
    ax.set_xlabel(xl)
    ax.set_ylabel(yl)
    ax.grid(True)
#     ax.set_yscale('log')
    for n,t in enumerate(d):
        ax.text(x[n],y[n],str(t-1),fontsize=7)
    
```

```python
fig,axs=plt.subplots(1,2,figsize=(8,4),dpi=200)
plt.subplots_adjust(hspace=0.3,wspace=0.35)
xvs=['Rmax0']
yvs=['thalf_PD','thalf_V10max']

xlabs=['Rmax$_{LF}$ (km)']
ylabs=['t$_{1/2}$ (h)']

#         plot_dec(axs[j,i],df0w[xvs[i]].to_numpy(),df0w[yvs[j]].to_numpy(),df0w['day'].to_numpy(),xlabs[i],ylabs[j])
plot_dec(axs[0],df1w[xvs[0]].to_numpy(),df1w[yvs[0]].to_numpy(),df1w['day'].to_numpy(),xlabs[0],ylabs[0])      
plot_dec(axs[0],df1w[xvs[0]].to_numpy(),df1w[yvs[1]].to_numpy(),df1w['day'].to_numpy(),xlabs[0],ylabs[0])
plot_dec(axs[1],df1w[xvs[0]].to_numpy(),df1d[yvs[0]].to_numpy(),df1d['day'].to_numpy(),xlabs[0],ylabs[0])      
plot_dec(axs[1],df1w[xvs[0]].to_numpy(),df1d[yvs[1]].to_numpy(),df1d['day'].to_numpy(),xlabs[0],ylabs[0])
axs[0].set_title('Wet')
axs[1].set_title('Dry')

# axs[0,0].legend(('run00','run01','run01dry'));
axs[0].legend(('Pmin','Vmax'));
```
