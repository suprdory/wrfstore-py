```python
import os, glob
from matplotlib import pyplot as plt
import pandas as pd
import importlib
import wrftools as wrf
importlib.reload(wrf)
```

```python
def save_run_max_df(flist,run,rs):
    x=[(print(f) or wrf.wrf2r(run,f,'P',rs,minim=False)) for n, f in enumerate(flist)]
    p10,p20,p30,p40,p50,p60,p70,p80,p90,p100,p150,p200,t=zip(*x)
    dfpr = pd.DataFrame(zip(p10,p20,p30,p40,p50,p60,p70,p80,p90,p100,p150,p200), index=t, columns=['p10','p20','p30','p40','p50','p60','p70','p80','p90','p100','p150','p200'])

    x=[(print(f) or wrf.wrf2max(run,f,'V')) for n, f in enumerate(flist)]
    vmax,rmax,zmax,t=zip(*x)
    dfVmax = pd.DataFrame(zip(vmax, rmax), index=t, columns=['V10max','rV10max'])

    x=[(print(f) or wrf.wrf2max(run,f,'vt')) for n, f in enumerate(flist)]
    vmax,rmax,zmax,t=zip(*x)
    dfvtmax = pd.DataFrame(zip(vmax, rmax), index=t, columns=['vt10max','rvt10max'])

    x=[(print(f) or wrf.wrf2max(run,f,'P',minim=True)) for n, f in enumerate(flist)]
    pmin,prmax,pzmax,t=zip(*x)
    dfpmin = pd.DataFrame(pmin, index=t, columns=['pmin'])
    
    df=dfvtmax.merge(dfVmax,how='outer',left_index=True,right_index=True)
    df=df.merge(dfpmin,how='outer',left_index=True,right_index=True)
    df=df.merge(dfpr,how='outer',left_index=True,right_index=True)

    sname='rundf/df_r_' + run + '.pkl'
    df.to_pickle(sname,protocol=4)
```
