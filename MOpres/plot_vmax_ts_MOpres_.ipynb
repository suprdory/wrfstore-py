```python
from matplotlib import pyplot as plt
import pandas as pd
plt.rcParams['figure.facecolor'] = 'white'
```

```python
df0=pd.read_pickle("./df_run_CTRL_.pkl")

runs=['03','04','05','06','07','08','09','10','11']
runs=['06']
fnames=['df_run_land_t' + r + '_.pkl' for r in runs]
dfs=[pd.read_pickle(f) for f in fnames]
```

```python
fig,axs=plt.subplots(3, 1, sharex='col')

linw=2

tCT=(df0.index-5)*24
axs[0].plot(tCT,df0.V10max,color='C0',lw=linw,linestyle='-')
axs[1].plot(tCT,df0.rV10max,color='C1',lw=linw,linestyle='-')
axs[2].plot(tCT,df0.pmin,color='C2',lw=linw,linestyle='-')

axs[0].set_ylabel('V$_{max}$ (m s$^{-1}$)',color='C0')
axs[1].set_ylabel('R$_{max}$ (km)',color='C1')
axs[2].set_ylabel('P$_{min}$ (hPa)',color='C2')

linw=2
for n,df in enumerate(dfs):
    tEX=(df.index-5)*24
    axs[0].plot(tEX,df.V10max,color='C0',lw=linw,linestyle='--')
    axs[1].plot(tEX,df.rV10max,color='C1',lw=linw,linestyle='--')
    axs[2].plot(tEX,df.pmin,color='C2',lw=linw,linestyle='--')

# axs[0].set_xlabel('Time (d)')
# axs[0].set_position
xmin=-3
xmax=14
axs[0].set_xlim(xmin,xmax)
# plt.xticks(range(xmin,round(xmax)))
axs[0].grid(True)

# axs[1].set_xlabel('Time (d)')
# axs[1].set_position
axs[1].grid(True)

axs[2].set_xlabel('Time after landfall (h)')
axs[2].set_position
axs[2].grid(True)

axs[1].set_ylim(10,30)
axs[2].set_ylim(895,990)

plt.sca(axs[0])
plt.legend(['CTRL','LAND'])

fig.set_size_inches(6,6)
fig.savefig('vmax_ts_MOpres.png',dpi=300,bbox_inches='tight', pad_inches=0.1,transparent=False)
plt.show()
```
