```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = 'white'
```

```python
N=100
X=1
x=np.linspace(0,1,N)
# y0=10

lam=X

# k=(2*np.pi)/lam
# y1=np.sin(k*x)
# y=y0+y1
# y=y-np.mean(y)


def get_fft(y):
    return(np.abs(np.fft.fft(y))[:6])

def get_yk(k):
    if k==0:
        return(x*0+1)
    else:
        return(np.sin(k*2*np.pi*x))
```

```python
ks=range(5)
fig,axs=plt.subplots(2,5,dpi=200,figsize=(10,4))
for n,k in enumerate(ks):
    if k==0:
        lam=np.inf
    else:
        lam=1/(k)
    axs[0,n].plot(x,get_yk(k))
    axs[0,n].set_title('k='+str(k)+', $\lambda$=' + '{:.2f}'.format(lam))
    axs[1,n].plot(get_fft(get_yk(k)))
    axs[1,n].set_xticks([0,1,2,3,4,5])
    axs[0,n].set_xticks([0,1])
    axs[0,n].set_yticks([-1,0,1])
    axs[1,n].set_yticks([0])
axs[0,0].set_ylabel('x')
axs[1,0].set_ylabel('abs(fft(x))')
```
