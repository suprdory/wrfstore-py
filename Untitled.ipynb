```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R=5
x=np.linspace(0,R,100)

A=x*np.sqrt(R**2-x**2)+R**2*np.arctan(x/np.sqrt(R**2-x**2))
Af=A/(np.pi*R**2)
```

```python
plt.plot(x,Af)
```

```python
p=np.linspace(1,0,100)
A=1
C=-1
a=1/p
b=np.log(1-p)
c=-np.log(p)

t=(a+b+c-C)/A
ta=(a-C)/A
tb=(b-C)/A
tc=(c-C)/A

plt.plot(t,p)
plt.plot(ta,p)
plt.plot(tb,p)
plt.plot(tc,p)

plt.xlim(-1,20)
plt.legend(('t','ta','tb','tc'))

# t=np.linspace(0,20,100)
# pa=1/(A*t+C)
# plt.plot(t,pa)
```
