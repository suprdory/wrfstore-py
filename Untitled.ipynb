```python
from matplotlib import pyplot as plt
from matplotlib import cm
import pandas as pd
plt.rcParams['figure.facecolor'] = 'white'
```

```python
#df0=pd.read_pickle("./df_run_CTRL_.pkl")

runs=['03','04','05','06','07','08','09','10','11']
fnames=['df_run_land_t' + r + '_.pkl' for r in runs]
dfs=[pd.read_pickle(f) for f in fnames]
```

```python
fig,axs=plt.subplots(3, 1, sharex='col')

# axs[0].set_ylabel('V$_{max}$ (m s$^{-1}$)',color='C0')
# axs[2].set_ylabel('R$_{max}$ (km)',color='C1')
# axs[1].set_ylabel('P$_{min}$ (hPa)',color='C2')

Vmax0=[]
Rmax0=[]
for n,df in enumerate(dfs):
    Vmax0.append(df.V10max.values[0])
    Rmax0.append(df.rV10max.values[0])
    
seq = sorted(Vmax0)
VmaxRank = [seq.index(v) for v in Vmax0]

seq = sorted(Rmax0)
RmaxRank = [seq.index(v) for v in Rmax0]

axs[0].set_ylabel('V$_{max}$/V$_{max}$lf')
axs[2].set_ylabel('R$_{max}$ (km)')
axs[1].set_ylabel('P$_{min}$/P$_{min}$lf')

linw=1.5
maxt=24

cmap=cm.get_cmap('viridis',len(dfs))

for n,df in enumerate(dfs):
    v10=df.V10max.values[:maxt]
    rv10=df.rV10max.values[:maxt]
    pmin=df.pmin.values[:maxt]
    
    v10=v10/v10[0]
    pmin=(pmin-1016)/(pmin[0]-1016)
    
    
    axs[0].plot(v10,color=cmap(RmaxRank[n]),lw=linw,linestyle='-')
    axs[2].plot(rv10,color=cmap(RmaxRank[n]),lw=linw,linestyle='-')
    axs[1].plot(pmin,color=cmap(RmaxRank[n]),lw=linw,linestyle='-')

xmax=maxt
axs[0].set_xlim(0,xmax)
plt.xticks(range(0,xmax,2))
axs[0].grid(True)

axs[1].grid(True)

axs[2].set_xlabel('Time (d)')
axs[2].set_position
axs[2].grid(True)

fig.suptitle('Rank by Rmax0')
fig.set_size_inches(10,10)
fig.savefig('decay_comp_norm_rankRmax.png',dpi=300,bbox_inches='tight', pad_inches=0.1,transparent=False)
plt.show()
```
