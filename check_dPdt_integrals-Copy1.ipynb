```python
import numpy as np
import matplotlib.pyplot as plt

Ph0=0.1
a=15000
Vmax=50
Rmax=30000

alp=2*a/(Vmax**2*Rmax)
```

```python
# (np.log(P)-np.log(P-Pe))/Pe=alp*t+C

dt=60
maxt=60*60*12
t=np.linspace(0,maxt,round(maxt/(dt*1)))
thr=t/60/60
Ph=np.zeros(len(t))
Ph[0]=Ph0

Pha=np.zeros(len(t))
Pha[0]=Ph0
# get numerical integral
for i in range(len(t)-1):
    dPhdt=-alp*(Ph[i]**2-Ph[i]**3)
    dPh=dPhdt*dt
    Ph[i+1]=Ph[i]+dPh
    
    dPhadt=-alp*(Pha[i]**2)
    dPha=dPhadt*dt
    Pha[i+1]=Pha[i]+dPha

# approx analytic integral (ignore complex log)
Phi=Ph0*np.exp(-Ph*alp*t)
```

```python
plt.plot(thr,Ph)
plt.plot(thr,Pha)
plt.grid(True)
```
