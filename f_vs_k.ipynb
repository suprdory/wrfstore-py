```python
import matplotlib.pyplot as plt
import numpy as np

def f(k):
    return((1-2**(1-k))/(2**(1-k)*(k-1)))
def f1(k):
    return((2**(k-1)-1)/(k-1))

k=np.linspace(0.5,2.5,1000)


fig,ax=plt.subplots(1,1,dpi=200)
ax.plot(k,f(k))
# plt.plot(k,f1(k),'--')
ax.grid()
ax.set_xlabel('k')
ax.set_ylabel(r'$\frac{2^{(k-1)}-1}{k-1}$')
```
