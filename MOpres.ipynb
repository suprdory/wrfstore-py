```python
from matplotlib import pyplot as plt
import numpy as np
import importlib
import wrftools as wrf
importlib.reload(wrf)

run='run_CTRL'
runex='run_land_t06'
d=3
flist=wrf.getflist(run=run,d=d)
f=flist[5*24+3]

th=wrf.getWRF(run,f,'th',z=0,type='az')
thex=wrf.getWRF(runex,f,'th',z=0,type='az')

r=wrf.getRcoord(wrf.getpath(run,f))
```

```python
plt.figure(figsize=(6, 4), dpi=200, facecolor='w', edgecolor='k')
plt.plot(r,th)
plt.plot(r,thex)
plt.ylim(0,50)
plt.xlim(0,200)
plt.xlabel('R (km)')
plt.ylabel('Inflow Angle ($^\circ$)')
plt.grid()
plt.legend(['CTRL +3hr','LAND +3hr'])
plt.show
plt.savefig('inflow_angle.png',bbox_inches='tight', pad_inches=0.1,transparent=False)

```
