```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.facecolor'] = 'white'

Pe=101600
Pmin=10000

lam=25000
alp=0.8
f=10e-4
mu=np.sqrt(Pmin*2*alp)

r=np.linspace(1,150000,1000)
dr=r[1]-r[0]

eps=np.exp(-r**2/(2*lam**2))

dPdr=(2*mu**2*lam**2)*(1-eps)/(alp*r**3)-(eps*mu**2)/(alp*r)-(r*f**2)/(4*alp)
Pnum=np.cumsum(dPdr)*dr+(Pe-Pmin)


P=((mu**2)/(2*alp))*(1-eps)/(np.log(eps))-f**2*r**2/(8*alp)+Pe
Pp=((mu**2)/(2*alp))*(1-eps)/np.log(eps)+Pe

# plt.plot(r,dPdr)
# plt.plot(r,dPdr)
# plt.plot(r,Pnum/100)
fig,ax=plt.subplots(1,1,figsize=(3,3),dpi=200)
plt.plot(r/1000,P/100)
plt.plot(r/1000,Pp/100)
plt.legend(('P(r)','P\'(r)'))
plt.xlabel('R (Km)')
plt.ylabel('P (hPa)')
plt.grid(True)

# plt.plot(r,(0-eps))
# plt.plot(r,(1-eps)/np.log(eps))
# plt.plot(r,np.cumsum(dPdr)*dr)
```
